## Top of proteines

```python
#Listes des Best features pour les proteines
dict_protein = {"APAAC":[50],"CTDT_CTDC_CTDD":[50,75,100],"CTriad":[50,100,150,200,250,300],
                "Geary":[50,100,150,200],"Moreau":[50,100,150,200],"Moran":[50,100,150,200],
                "QSO":[50,75],"Dipetite":[50,100,200,300],"SOCN":[50],
                "Tripetite":[50,100,200,1000,2000,3000,3500,4000,4500]}
```

```python
for i in range(len(org_results)):
    path_org = "./Results/"+org_results[i]+"/protein_subgroups/feat_importance"
    path_collection = "./Features_collections/"+org_groups[i]+"/subgroup_protein"
    file_importance = ""
    file_collection = ""
    path_top = "./Features_collections/"+org_groups[i]+"/Top_Proteines_features"
    os.mkdir(path_top)
    !cp $script $path_top
    for key, value in dict_protein.items():
        for fichier in os.listdir(path_org):
            x = fichier.split("_F")
            if x[0].lower() == key.lower():
                #file_importance = path_org+"/"+fichier
                f = path_org.split(".")
                file_importance = "../../.."+f[1]+"/"+fichier
        for fichier in os.listdir(path_collection):
            x = fichier.split("n_")
            y = x[1].split(".")
            if y[0].lower() == key.lower():
                file_collection = "../subgroup_protein/"+fichier
        for val in value:
            %cd $path_top
            !pwd
            !Rscript $script $file_importance $file_collection $val $key
            %cd $parent_path
            
                
    
```
