```python
# Data set
url = '/content/dendrogramData.csv'
df = pd.read_csv(url,sep=";")
df = df.replace(np.nan,0)
df = df.set_index('Feature')
model= df.pop("Model")
print(model)
df
```

```python
# Prepare a vector of color mapped to the 'cyl' column
#my_palette = dict(zip(df.Model.unique(), ["orange","yellow","brown","blue","black","green","red"]))
lut = dict(zip(model.unique(), ["orange","aqua","brown","blue","fuchsia","green","purple"])) #"rbg"
row_colors =model.map(lut)
#print(lut)
# plot
graph=sns.clustermap(df,figsize=(14, 11), row_colors=row_colors)
for key, values in lut.items():
  graph.ax_col_dendrogram.bar(0, 0, color=values,
                            label=key, linewidth=0)
graph.ax_col_dendrogram.legend(loc="right", ncol=1)
#sns.clustermap(df, metric="correlation", method="single", cmap="Blues", standard_scale=1, row_colors=row_colors)  cmap="mako", vmin=0, vmax=10
#figsize=(7, 15), row_cluster=False, dendrogram_ratio=(.1, .2), cbar_pos=(0, .2, .03, .4)
plt.show()
```
