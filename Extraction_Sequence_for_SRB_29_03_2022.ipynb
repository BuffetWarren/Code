#Extraction des sequences d'acide amine et Nucleotides pour les 25 Organismes SRB

## Principe:



*   on ouvre le fichier **SRB Analyst dataset-org.csv** qui contient les 25 organismes
*   On insere les abreviation et noms d'organismes dans deux listes qu'on utilisera durant tout le programme
*   Par L'url on accede on different elements de cgaque organisme
*   Lorsque nous avons la liste des elements d'un organisme on extrait L'id qu'on ajoute a url_get afin d'avoir les sequences voulu(aa ou nt)

## Resultat:
***Pour chaque organisme on aura un dossier contenant les sequences aa et nt pour cet organisme***

```python
from pandas import *
 
data = read_csv("SRB Analyst dataset-org.csv")
```

```python
#on recupere les noms ainsi que les abreviations d'organismes
#On les garde dans deux listes
abr_name = data['Abbr_Name'].tolist()
organism_name = data['Organism name'].tolist()
```

```python
!pip install biopython
from Bio import SeqIO
import requests
```

```python
#url pour accceder a la liste des organismes
url = "http://rest.kegg.jp/list/"
#url pour accceder a tous les elements d'un organisme
url_get = "http://rest.kegg.jp/get/"
path = "SRB_sequences/"
#for i in range(len(abr_name)-23): ## ce que j'ai executer chez moi pour 
for i in range(16,len(abr_name)):
    ###On cree un repertoire pour chaque organisme avec son son###
    val = organism_name[i]
    x = val.split()
    prem = x[0]+"_"
    deu = x[1]
    rep = prem+deu
    # creation du repertoire de l'organisme i
    !mkdir $path$prem$deu
    response = requests.get(url+abr_name[i])
    list_organism = response.text

    ###Sauvegarde des lement de l'organisme i dans son repertoire##
    with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_elements.csv","a") as f:
        for line in list_organism:
            f.write("%s" %line)

    listes = open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_elements.csv", "r")
    list_id = []

    ##On fait le split pour avoir uniquement les id##
    for line in listes:
      x = line.split()
      list_id.append(x[0])
    
    ##Bloc de recuperation des sequences##
    #for k in range(len(list_id)-1500): ## ce que j'ai executer chez moi
    for k in range(len(list_id)):
      #A partir de l'url url_get on ouvre pour chaque id on recupere la sequence dans le cas des aa ou nt#
      response_aa = requests.get(url_get+list_id[k]+"/aaseq")
      list_aa = response_aa.text
      response_nt = requests.get(url_get+list_id[k]+"/ntseq")
      list_nt = response_nt.text
      #Sauvegarde de la sequence aa dans le repertoire i et fichier correspondant#
      with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_aaseq.fasta", 'a') as temp_file:
        for item in list_aa:
          temp_file.write("%s" % item)
      #Sauvegarde de la sequence nt dans le repertoire i et fichier correspondant#
      with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"ntseq.fasta", 'a') as temp_file:
        for item in list_nt:
          temp_file.write("%s" % item)
```

Nettoyage des fastas

```python
#Nettoyage des acides aa
for i in range(len(abr_name)):
  val = organism_name[i]
  x = val.split()
  prem = x[0]+"_"
  deu = x[1]
  rep = prem+deu
  with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_aaseq.fasta") as handle:
    with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_aaseq_clean.fasta","a") as clean_file:
      for record in SeqIO.parse(handle, "fasta"):
        clean_file.write('>'+"%s" %record.id.upper()+'\n')
        clean_file.write("%s" %record.seq.upper()+'\n')
```

```python
#Nettoyage des acides nt
for i in range(len(abr_name)):
  val = organism_name[i]
  x = val.split()
  prem = x[0]+"_"
  deu = x[1]
  rep = prem+deu
  with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_ntseq.fasta") as handle:
    with open("SRB_sequences/"+rep+"/SRB_"+abr_name[i]+"_ntseq_clean.fasta","a") as clean_file:
      for record in SeqIO.parse(handle, "fasta"):
        clean_file.write('>'+"%s" %record.id.upper()+'\n')
        clean_file.write("%s" %record.seq.upper()+'\n')
```
